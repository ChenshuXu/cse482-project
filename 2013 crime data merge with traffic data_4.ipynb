# 2013 crime data merge with traffic data

```python
import pandas as pd
```

```python
crime_data = pd.read_pickle("crime_data_2013.pkl")
crime_data
```

```python
traffic_data = pd.read_pickle("traffic_data_in_segments.pkl")
traffic_data
```

```python
crime_data.shape[0]
```

```python
total = crime_data.shape[0]
start = int(total * (3/4))
end = total-1
find_lst = []
for crime_index in range(start,end):
    crime_row = crime_data.loc[crime_index]
    for traffic_index, traffic_row in traffic_data.iterrows():
        if crime_row['Latitude'] >= traffic_row['START_LATITUDE'] \
        and crime_row['Latitude'] <= traffic_row['END_LATITUDE'] \
        and crime_row['Longitude'] >= traffic_row['START_LONGITUDE'] \
        and crime_row['Longitude'] <= traffic_row['END_LONGITUDE']:
            row = crime_row.tolist()
            row.extend(traffic_row.tolist())
            find_lst.append(row)
```

```python
df = pd.DataFrame(find_lst)
df.columns=['index', 'Date', 'Primary Type', 'District', 'Community Area', 'Latitude',
       'Longitude', 'Time Period', 'SPEED', 'Average Traffic condition', 'SEGMENTID', 'START_LONGITUDE',
       'START_LATITUDE', 'END_LONGITUDE', 'END_LATITUDE']
df
```

```python
df.to_pickle("crime_merge_traffic_2013_4.pkl")
```
