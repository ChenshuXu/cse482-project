```python
import pandas as pd
import operator
import datetime

data1 = pd.read_csv('Census_Data_-_Selected_socioeconomic_indicators_in_Chicago__2008___2012.csv')

data1.columns

data1_sub = data1[ ['Community Area Number', 'COMMUNITY AREA NAME', 'PER CAPITA INCOME '] ]

data1_sub = data1_sub.sort_values(by = ['PER CAPITA INCOME '])

data1_sub = data1_sub.reset_index(drop=True)

data1_sub
```

```python
#use equal width
# max_inc = max(data1_sub['PER CAPITA INCOME '])
# min_inc = min(data1_sub['PER CAPITA INCOME '])
# interval = (max_inc - min_inc)/ 3

#use equal frequency
count = data1_sub.shape[0] - 1

#get the split points
point1 = int(count/3)
point2 = int(2 * count/3)

income_subData = pd.DataFrame(columns=['Community Area Number', "PER CAPITA INCOME"]);

for i in range(0, count + 1):
    
    com_area = data1_sub.iloc[i]['Community Area Number']
    
    if data1_sub.iloc[i]['COMMUNITY AREA NAME'] != "CHICAGO":
        if i < point1:
            income_subData.loc[i] = (com_area, "Low")
        elif i < point2:
            income_subData.loc[i] = (com_area, "Medium")
        else:
            income_subData.loc[i] = (com_area, "High")
    
income_subData.groupby('PER CAPITA INCOME').count()
```

```python
income_subData = income_subData.sort_values(by = ['Community Area Number'])

income_subData = income_subData.reset_index(drop=True)

income_subData
```

```python
income_subData.to_pickle("Income_subData.pkl")
```

```python
data13_14 = pd.read_csv('Weather/2013_01_01_to_2014_12_31.csv')

data13_14 = data13_14.sort_values(by = ['DATE'])

data13_14 = data13_14[ ['NAME', 'DATE', 'WT01', 'WT02', 'WT03', 'WT04', 'WT05', 'WT06', 'WT08', 'WT09', 'WT10', 'WT11', 'WT13', 'WT14', 'WT16', 'WT17', 'WT18', 'WT19', 'WT22'] ]

data13_14 = data13_14.reset_index(drop=True)
```

```python
#weather type data for 2013

data13_result = pd.DataFrame(columns=['DATE', "WEATHER", "LOCATION"]);

result_index = 0;
        
for i in data13_14.index:
    
    date = data13_14.iloc[i]['DATE']
    
    location = data13_14.iloc[i]['NAME']
    
    if "IL US" in location and date[:4] == "2013":

        # check if the weather is Snow
        if data13_14.iloc[i]['WT18'] == 1 or data13_14.iloc[i]['WT09'] == 1 or data13_14.iloc[i]['WT04'] == 1 or data13_14.iloc[i]['WT05'] == 1 or data13_14.iloc[i]['WT06'] == 1 or data13_14.iloc[i]['WT02'] == 1:

            data13_result.loc[result_index] = (date, "Snow", location)

        # check if the weather is Rain
        elif data13_14.iloc[i]['WT14'] == 1 or data13_14.iloc[i]['WT16'] == 1 or data13_14.iloc[i]['WT17'] == 1 or data13_14.iloc[i]['WT19'] == 1:

            data13_result.loc[result_index] = (date, "Rain", location)

        # check if the weather is Cloudy
        elif data13_14.iloc[i]['WT03'] == 1 or data13_14.iloc[i]['WT08'] == 1 or data13_14.iloc[i]['WT10'] == 1 or data13_14.iloc[i]['WT11'] == 1 or data13_14.iloc[i]['WT13'] == 1 or data13_14.iloc[i]['WT22'] == 1:

            data13_result.loc[result_index] = (date, "Cloudy", location)

        # check if the weather is Sunny
        else:

            data13_result.loc[result_index] = (date, "Sunny", location)

        result_index += 1;
        
print(data13_result.groupby('WEATHER').count())
print(data13_result.groupby('LOCATION').count())
```

```python

data13_subData = pd.DataFrame(columns=['DATE', "WEATHER"]);

sub_index = 0;

for date, wt in data13_result.groupby('DATE'):
    weather_type = {'Sunny':0, 'Cloudy':0, 'Rain':0, 'Snow': 0}
    sunny = 0
    cloud = 0
    rain = 0
    snow = 0
    for weather in wt['WEATHER']:
        if weather == "Sunny":
            weather_type['Sunny'] += 1
        elif weather == "Cloudy":
            weather_type['Cloudy'] += 1
        elif weather == "Rain":
            weather_type['Rain'] += 1
        elif weather == "Snow":
            weather_type['Snow'] += 1
    
    weather = max(weather_type.items(), key=operator.itemgetter(1))[0]
    
    max_val = max(weather_type.items(), key=operator.itemgetter(1))[1]
    
    if weather_type['Snow'] == weather_type[weather]:
        weather = 'Snow'
    
    elif weather_type['Rain'] == weather_type[weather]:
        weather = 'Rain'
    
    elif weather_type['Cloudy'] == weather_type[weather]:
        weather = 'Cloudy'
    
    elif weather_type['Sunny'] == weather_type[weather]: 
        weather = 'Sunny'
    
    
    data13_subData.loc[sub_index] = (date, weather)
    sub_index += 1

print(data13_subData.groupby('WEATHER').count())
```

```python

data13_finalData = pd.DataFrame(columns=['DATE', "WEATHER"]);

sub_date = pd.to_datetime("2013-01-01")

index = 0

for i in range(1, 366) :
    if data13_subData.iloc[index]['DATE'] != sub_date.strftime('%Y-%m-%d'):
        data13_finalData.loc[i] = (sub_date.strftime('%Y-%m-%d'), "Sunny")
    else:
        data13_finalData.loc[i] = data13_subData.iloc[index]
        index += 1
    sub_date = sub_date + datetime.timedelta(days=1)
    
data13_finalData
```

```python
print(data13_finalData.groupby('WEATHER').count())
```

```python
data13_finalData.to_pickle("Weather_Type_13_subData.pkl")
```
