```python
import pandas as pd
```

```python
crime_data = pd.read_pickle("crime_data_2014.pkl")
crime_data
```

```python
day = []

for index,row in crime_data.iterrows():
    day.append(row['Datetime'].strftime("%Y-%m-%d"))
```

```python
crime_data['Day'] = day
```

```python
crime_count_in_district_each_day = crime_data.groupby(by=['Day','District']).count()
crime_count_in_district_each_day
```

```python
total_crime_each_day = crime_data.groupby(by=['Day']).count()
total_crime_each_day
```

```python
crime_rate = []

for index, row in crime_count_in_district_each_day.iterrows():
    day = index[0]
    total_row = total_crime_each_day.loc[day]
    total = total_row['District']
    count = row['Primary Type']
    rate = count/total
    district = index[1]
    new_row = [day, district, rate]
    crime_rate.append(new_row)

crime_rate
```

```python
crime_rate_df = pd.DataFrame(crime_rate, columns=['Day','District','Crime Rate'])
crime_rate_df
```

```python
crime_rate_df.to_pickle("2014 crime_rate.pkl")
```
