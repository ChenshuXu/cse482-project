# Merge 4 fragments together

```python
import pandas as pd
```

```python
crime_data_1 = pd.read_pickle("crime_merge_traffic_2013_1.pkl")
```

```python
crime_data_2 = pd.read_pickle("crime_merge_traffic_2013_2.pkl")
```

```python
crime_data_3 = pd.read_pickle("crime_merge_traffic_2013_3.pkl")
```

```python
crime_data_4 = pd.read_pickle("crime_merge_traffic_2013_4.pkl")
```

```python
new_df = crime_data_1.append(crime_data_2)
```

```python
new_df = new_df.append(crime_data_3)
```

```python
new_df = new_df.append(crime_data_4)
```

```python
new_df = new_df.reset_index()
```

```python
new_df = new_df.drop(columns=['level_0','index','START_LONGITUDE','START_LATITUDE','END_LONGITUDE', 'END_LATITUDE'])
```
