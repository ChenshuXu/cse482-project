# 2014 crime data merge with traffic data

Load crime data

```python
import pandas as pd
```

```python
crime_data = pd.read_pickle("crime_data_2014.pkl")
crime_data
```

```python
crime_data.columns
```

Load traffic data

```python
traffic_data = pd.read_pickle("traffic_data_in_segments.pkl")
traffic_data
```

```python
traffic_data.columns
```

```python
find_lst = []
for crime_index, crime_row in crime_data.iterrows():
    for traffic_index, traffic_row in traffic_data.iterrows():
        if crime_row['Latitude'] >= traffic_row['START_LATITUDE'] \
        and crime_row['Latitude'] <= traffic_row['END_LATITUDE'] \
        and crime_row['Longitude'] >= traffic_row['START_LONGITUDE'] \
        and crime_row['Longitude'] <= traffic_row['END_LONGITUDE']:
            row = crime_row.tolist()
            row.extend(traffic_row.tolist())
            find_lst.append(row)
```

```python
df = pd.DataFrame(find_lst)
```

```python
df.columns=['ID', 'Date', 'Block', 'Primary Type', 'District', 'Community Area',
       'Latitude', 'Longitude', 'Time Period', 'Datetime','SPEED', 'Traffic condition', 'SEGMENTID', 'START_LONGITUDE',
       'START_LATITUDE', 'END_LONGITUDE', 'END_LATITUDE']
```

```python
df
```

```python
new_df = df.drop(columns=['ID','Date','Block','START_LONGITUDE',
       'START_LATITUDE', 'END_LONGITUDE', 'END_LATITUDE'])
new_df = new_df.rename(columns={'Traffic condition':'Average Traffic condition'})
```
