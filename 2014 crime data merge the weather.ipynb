# 2014 crime data merge the weather data

```python
import pandas as pd

crime_data = pd.read_pickle("crime_merge_traffic_2014.pkl")
crime_data
```

```python
crime_data.columns
```

```python
weather_data = pd.read_pickle("Weather_Type_14_subData.pkl")
weather_data
```

#### Merge crime data with weather data on date

```python
crime_data["DATE"] = crime_data["Datetime"].dt.strftime('%Y-%m-%d')
```

```python
crime_data
```

```python
merged_data = pd.merge(crime_data, weather_data, left_on='DATE', right_on='DATE', how='inner')
```

```python
merged_data = merged_data.drop(columns=['DATE'])
merged_data
```

```python
merged_data.to_pickle("crime_data_traffic_weather_2014.pkl")
```
