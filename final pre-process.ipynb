```python
import pandas as pd
```

```python
crime_data_2013 = pd.read_pickle("data_final_2013.pkl")
crime_data_2013
```

```python
crime_data_2014 = pd.read_pickle("data_final_2014.pkl")
crime_data_2014
```

```python
crime_data = crime_data_2013.append(crime_data_2014)
crime_data = crime_data.reset_index()
crime_data = crime_data.drop(columns=['index'])
crime_data
```

```python
traffic_speed = pd.read_pickle("traffic_data_in_segments.pkl")
traffic_speed
```

```python
crime_data2 = pd.merge(crime_data,traffic_speed,left_on='SEGMENTID',right_on='SEGMENTID')
crime_data2 = crime_data2.drop(columns=['Average Traffic condition_y','START_LONGITUDE','START_LATITUDE','END_LONGITUDE','END_LATITUDE'])
crime_data2
```

```python
income_data = pd.read_csv("Census_Data_-_Selected_socioeconomic_indicators_in_Chicago__2008___2012.csv")
income_data = income_data.dropna()
income_data = income_data[['Community Area Number','PER CAPITA INCOME ']]
income_data
```

```python
crime_data3 = pd.merge(crime_data2,income_data,left_on='Community Area',right_on='Community Area Number')
crime_data3 = crime_data3.drop(columns=['Community Area Number'])
crime_data3
```

```python
crime_data3.to_pickle('final_data.pkl')
```

```python
crime_data3.to_csv("final_data.csv")
```
