```python
import pandas as pd
```

## Process 2013 data

```python
crime_data_2013 = pd.read_pickle("crime_data_traffic_weather_income_2013.pkl")
crime_data_2013
```

```python
day = []

for index,row in crime_data_2013.iterrows():
    day.append(row['Date'].strftime("%Y-%m-%d"))

crime_data_2013['Day'] = day
crime_data_2013
```

```python
crime_rate_2013 = pd.read_pickle("2013 crime_rate.pkl")
crime_rate_2013
```

```python
df_2013 = pd.merge(crime_data_2013,crime_rate_2013,left_on=['District','Day'],right_on=['District','Day'])
df_2013
```

```python
dayOfWeek = []

for index,row in df_2013.iterrows():
    day = row['Date'].strftime("%a")
    if day == 'Sun' or day == 'Sat':
        dayOfWeek.append('weekend')
    else:
        dayOfWeek.append('weekday')
dayOfWeek

df_2013['dayOfWeek'] = dayOfWeek
```

```python
df_2013.columns
```

```python
store_2013 = df_2013.drop(columns=['Primary Type', 'Latitude',
       'Longitude','SPEED'])
store_2013
```

```python
store_2013.to_pickle("data_final_2013.pkl")
```

## Process 2014 data

```python
crime_data_2014 = pd.read_pickle("crime_data_traffic_weather_income_2014.pkl")
crime_data_2014
```

```python
day = []

for index,row in crime_data_2014.iterrows():
    day.append(row['Datetime'].strftime("%Y-%m-%d"))

crime_data_2014['Day'] = day
crime_data_2014
```

```python
crime_rate_2014 = pd.read_pickle("2014 crime_rate.pkl")
crime_rate_2014
```

```python
df_2014 = pd.merge(crime_data_2014,crime_rate_2014,left_on=['District','Day'],right_on=['District','Day'])
df_2014
```

```python
dayOfWeek = []

for index,row in df_2014.iterrows():
    day = row['Datetime'].strftime("%a")
    if day == 'Sun' or day == 'Sat':
        dayOfWeek.append('weekend')
    else:
        dayOfWeek.append('weekday')

df_2014['dayOfWeek'] = dayOfWeek
```

```python
store_2014 = df_2014.drop(columns=['Primary Type', 'Latitude',
       'Longitude','SPEED'])

store_2014 = store_2014.rename(columns={'Datetime':'Date'})
store_2014
```

```python
store_2014.to_pickle("data_final_2014.pkl")
```
